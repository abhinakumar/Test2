## Q3

```python
def calculate_tax(*salaries):
    try:
        taxes = list(map(lambda sal: sal * 0.10, salaries))
        total_tax=sum(taxes)
        return total_tax

    except TypeError:
        print("error: salary must be numeric.")
    except Exception as e:
        print("error:", e)

print("Total tax =", calculate_tax(30000, 40000, 50000))
```

## Q5

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

print("First 10 Fibonacci numbers using loop :")
for num in fibonacci(10):
    print(num, end=" ")
print()
print("\nFirst 4 Fibonacci numbers using next():")
fib_gen = fibonacci(10)
print(next(fib_gen))
print(next(fib_gen))
print(next(fib_gen))
print(next(fib_gen))
```

## Q6

```python
%%writefile student_utils.py
def get_grade(avg, pass_mark=40):
    if avg < pass_mark:
        return "F"
    elif avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= pass_mark:
        return "D"


def calculate_average(*marks):
    return sum(marks) / len(marks)
```

```python
import json
from student_utils import get_grade, calculate_average

students_data = []

print("Enter details for 3 students:\n")

for i in range(3):
    print(f"--- Student {i + 1} ---")
    name = input("Enter name: ")

    marks = []
    for subj in range(1, 4):
        while True:
            try:
                mark = float(input(f"Enter marks for subject {subj}: "))
                if mark < 0 or mark > 100:
                    print("Invalid marks. Must be between 0 and 100. Try again.")
                    continue
                marks.append(mark)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

    avg = calculate_average(*marks)
    grade = get_grade(avg)

    students_data.append({
        "name": name,
        "marks": marks,
        "average": round(avg, 2),
        "grade": grade
    })

with open("report.txt", "w") as report_file:
    for student in students_data:
        line = f"Name: {student['name']} | Average: {student['average']:.2f} | Grade: {student['grade']}\n"
        report_file.write(line)

with open("students.json", "w") as json_file:
    json.dump(students_data, json_file, indent=4)

print("\n--- Contents of report.txt ---")
with open("report.txt", "r") as report_file:
    print(report_file.read())

print("--- Contents of students.json ---")
with open("students.json", "r") as json_file:
    print(json_file.read())

print("\n--- report.txt read using readlines() ---")
with open("report.txt", "r") as report_file:
    lines = report_file.readlines()
    for idx, line in enumerate(lines, start=1):
        print(f"Line {idx}: {line.strip()}")
```
